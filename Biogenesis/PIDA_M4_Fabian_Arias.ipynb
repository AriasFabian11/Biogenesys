```python
#lectura del dataset mediante pandas
df = pd.read_csv('data_latinoamerica.csv')
```

```python
df.columns
```

```python
#lectura de los diferentes paises
df['location_key'].unique()
```

```python
#filtro del dataset por los paises indicados
paises_seleccionados = ['AR','CL','CO','MX','PE','BR']
data_latam = df[df['location_key'].isin(paises_seleccionados)]
```

```python
#cambio de tipo de la columna 'date' a datetime
data_latam['date'] = pd.to_datetime(data_latam['date'])
```

```python
#filtro de fechas posteriores a 2021-01-01
data_latam_fechas = data_latam[data_latam['date']>'2021-01-01']
```

```python
data_latam_fechas
```

```python
#confirmamos valores nulos
valores_nulos_data = data_latam_fechas.isnull().sum()[data_latam_fechas.isnull().sum()>0]
print(valores_nulos_data)
print(np.shape(data_latam_fechas))
```

```python
#relleno de valores nulos valores nulos
#relleno por la media
for pais in paises_seleccionados:
    data_latam_fechas['new_confirmed']=data_latam_fechas['new_confirmed'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['new_confirmed'].mean())
    data_latam_fechas['new_deceased']=data_latam_fechas['new_deceased'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['new_deceased'].mean())
    data_latam_fechas['average_temperature_celsius']=data_latam_fechas['average_temperature_celsius'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['average_temperature_celsius'].mean())
    data_latam_fechas['minimum_temperature_celsius']=data_latam_fechas['minimum_temperature_celsius'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['minimum_temperature_celsius'].mean())
    data_latam_fechas['maximum_temperature_celsius']=data_latam_fechas['maximum_temperature_celsius'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['maximum_temperature_celsius'].mean())
    data_latam_fechas['rainfall_mm']=data_latam_fechas['rainfall_mm'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['rainfall_mm'].mean())
    data_latam_fechas['relative_humidity']=data_latam_fechas['relative_humidity'].fillna(data_latam_fechas[data_latam_fechas['location_key']==pais]['relative_humidity'].mean())

#relleno con el valor anterior o siguiente
data_latam_fechas['cumulative_confirmed'] = data_latam_fechas.groupby('location_key')['cumulative_confirmed'].fillna(method='ffill')
data_latam_fechas['cumulative_deceased'] = data_latam_fechas.groupby('location_key')['cumulative_deceased'].fillna(method='ffill')
data_latam_fechas['cumulative_vaccine_doses_administered'] = data_latam_fechas.groupby('location_key')['cumulative_vaccine_doses_administered'].fillna(method='ffill') 
data_latam_fechas['cumulative_vaccine_doses_administered'] = data_latam_fechas.groupby('location_key')['cumulative_vaccine_doses_administered'].fillna(method='bfill') # rellenamos con el siguiente en caso de seguir nulo

#relleno con 0
data_latam_fechas['new_recovered'] = data_latam_fechas['new_recovered'].fillna(0)
data_latam_fechas['cumulative_recovered'] = data_latam_fechas['new_recovered'].fillna(0)
```

```python
#confirmamos nuevamente valores nulos
valores_nulos_data = data_latam_fechas.isnull().sum()[data_latam_fechas.isnull().sum()>0]
print(valores_nulos_data)
```

```python
#Guardar el nuevo df en un archivo csv
data_latam_fechas.to_csv('DatosFinalesFiltrado.csv',index=False)
```

```python
#aplicacion de for para generar estadisticas descriptivas
for i in data_latam_fechas.columns:
    print(i)
    print(data_latam_fechas[i].describe())
    print('--------')
```

```python
#Funcion para obtener Mediana, Varianza, Rango, Moda
def descripcion_datos(datos):
    for i in datos.columns:
        if datos[i].dtype=='float64' or datos[i].dtype=='int64':
            print(i)
            print('Media: ', datos[i].mean())
            print('Mediana: ', datos[i].median())
            print('Varianza: ', datos[i].var())
            print('Rango: ', datos[i].max()-datos[i].min())
            print('Moda: ', datos[i].mode().iloc[0]) #mode devuelve un dataframe
            print('--------------')

descripcion_datos(data_latam_fechas)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#leemos el archivo filtrado para liberar memoria
df = pd.read_csv('DatosFinalesFiltrado.csv')
```

```python
#EvoluciÃ³n de Casos Activos vs. Recuperados
# Convertir la columna 'date' a tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Filtrar datos para 2021 y 2022
datos_2021 = df[df['date'].dt.year == 2021]
datos_2022 = df[df['date'].dt.year == 2022]

# Agrupar por mes y calcular la suma de casos activos y recuperados
# Casos activos = nuevos confirmados - nuevos recuperados - nuevos fallecidos
datos_2021['casos_activos'] = datos_2021['new_confirmed'] - datos_2021['new_recovered'] - datos_2021['new_deceased']
datos_2022['casos_activos'] = datos_2022['new_confirmed'] - datos_2022['new_recovered'] - datos_2022['new_deceased']

# Agrupar por mes
datos_2021_mes = datos_2021.groupby(datos_2021['date'].dt.month).agg({
    'casos_activos': 'sum',
    'new_recovered': 'sum'}).reset_index()

datos_2022_mes = datos_2022.groupby(datos_2022['date'].dt.month).agg({
    'casos_activos': 'sum',
    'new_recovered': 'sum'
}).reset_index()

# Renombrar la columna de meses para que sea mÃ¡s clara
datos_2021_mes.rename(columns={'date': 'mes'}, inplace=True)
datos_2022_mes.rename(columns={'date': 'mes'}, inplace=True)

# Crear los grÃ¡ficos
plt.figure(figsize=(14, 6))

# GrÃ¡fico para 2021
plt.subplot(1, 2, 1)
sns.lineplot(data=datos_2021_mes, x='mes', y='casos_activos', label='Casos Activos', color='red')
sns.lineplot(data=datos_2021_mes, x='mes', y='new_recovered', label='Casos Recuperados', color='green')
plt.title('EvoluciÃ³n de Casos Activos vs. Recuperados (2021)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de casos')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()

# GrÃ¡fico para 2022
plt.subplot(1, 2, 2)
sns.lineplot(data=datos_2022_mes, x='mes', y='casos_activos', label='Casos Activos', color='red')
sns.lineplot(data=datos_2022_mes, x='mes', y='new_recovered', label='Casos Recuperados', color='green')
plt.title('EvoluciÃ³n de Casos Activos vs. Recuperados (2022)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de casos')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Nuevos casos y temperatura promedio
# Asegurarnos de que la columna 'date' estÃ© en formato datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupar por fecha y calcular la media mensual de 'new_confirmed' y 'average_temperature_celsius'
data_mensual = df.resample('ME', on='date').mean(numeric_only=True)

# Crear el grÃ¡fico
plt.figure(figsize=(20, 7))
ax1 = plt.gca()  # Primer eje y (nuevos casos confirmados)
ax2 = ax1.twinx()  # Segundo eje y (temperatura promedio)

# Graficar nuevos casos confirmados (lÃ­nea roja punteada)
ax1.plot(
    data_mensual.index, 
    data_mensual['new_confirmed'], 
    color='red', 
    label='Nuevos casos confirmados',
    linestyle=':',  # LÃ­nea punteada
    linewidth=2  # Hacer la lÃ­nea un poco mÃ¡s gruesa
)

# Graficar temperatura promedio (lÃ­nea azul continua)
ax2.plot(
    data_mensual.index, 
    data_mensual['average_temperature_celsius'], 
    color='blue', 
    label='Temperatura promedio (Â°C)', 
    linestyle='-',  # LÃ­nea continua
    linewidth=2  # Hacer la lÃ­nea un poco mÃ¡s gruesa
)

# Etiquetas y tÃ­tulo con fuentes mÃ¡s grandes
ax1.set_xlabel('Fecha', fontsize=14)
ax1.set_ylabel('Nuevos casos confirmados', color='red', fontsize=14)
ax2.set_ylabel('Temperatura promedio (Â°C)', color='blue', fontsize=14)
plt.title('Nuevos casos de COVID-19 y temperatura promedio con el tiempo', fontsize=16, pad=20)

# Leyendas con fuentes mÃ¡s grandes
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

# Ajustar el tamaÃ±o de las etiquetas de los ejes
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

# Mostrar el grÃ¡fico
plt.tight_layout()
plt.show()
```
